```python
#Merging database
DB_DOCK_final = pd.concat([HigDiv, Druglike], ignore_index=True)
DB_DOCK_final
```

```python
# Reference scores
NF2889=pd.read_csv("/content/drive/MyDrive/Cribado_Virtual_smHDAC8/Ledock_replicas/Sitio_alosterico_Ledock/DOCK_sitioAlosterico_NF2889.csv")
NF2889 = NF2889.rename(columns={'ID': 'ID_deNovo'})
NF2889
```

```python
# Average of the 3 replicates
NF2889['Docking_mean'] = NF2889[['Scores_Rep1', 'Scores_Rep2', 'Scores_Rep3']].mean(axis=1)

# Calculate pKd from the average
NF2889['pKd'] = (-NF2889['Docking_mean'] / 1.36).round(2)
NF2889
```

```python
# Merging
DB_DOCK = pd.concat([DB_DOCK_final, NF2889], join='inner', ignore_index=True)
DB_DOCK
```

```python
df_Rep1 = DB_DOCK[['ID_deNovo', 'Scores_Rep1']].copy()
df_Rep2 = DB_DOCK[['ID_deNovo', 'Scores_Rep2']].copy()
df_Rep3 = DB_DOCK[['ID_deNovo', 'Scores_Rep3']].copy()
df_Rep1
```

```python
# We extract the relevant columns into an array
Scores_Rep1 = df_Rep1['Scores_Rep1'].values
Scores_Rep2 = df_Rep2['Scores_Rep2'].values
Scores_Rep3 = df_Rep3['Scores_Rep3'].values
Scores_Rep1
```

```python
# Mean and standard deviation for Rep1
mu_rep1 = np.mean(Scores_Rep1)
sigma_rep1 = np.std(Scores_Rep1)

# Mean and standard deviation for Rep2
mu_rep2 = np.mean(Scores_Rep2)
sigma_rep2 = np.std(Scores_Rep2)

# Mean and standard deviation for Rep3
mu_rep3 = np.mean(Scores_Rep3)
sigma_rep3 = np.std(Scores_Rep3)
```

##Z-Score

$Z_j^i = \frac{1}{n} \sum_j\frac{s_j^i - \mu_j}{\sigma_j}$

```python
df = DB_DOCK[['ID_deNovo']].copy()

#  Z-scores for each method
df['Z_Rep1'] = (df_Rep1['Scores_Rep1'] - mu_rep1) / sigma_rep1
df['Z_Rep2'] = (df_Rep2['Scores_Rep2'] - mu_rep2) / sigma_rep2
df['Z_Rep3'] = (df_Rep3['Scores_Rep3'] - mu_rep3) / sigma_rep3

# Combined Z-score (average of the Z-scores)
df['Z_Score_Combined'] = (df['Z_Rep1'] + df['Z_Rep2'] + df['Z_Rep3']) / 3


# rank the combined Z-score
df["R_Z_Score"] = df["Z_Score_Combined"].rank(ascending=True, method="dense")

print("Estadísticas calculadas:")
print(f"REP1: μ={mu_rep1:.4f}, σ={sigma_rep1:.4f}")
print(f"REP2: μ={mu_rep2:.4f}, σ={sigma_rep2:.4f}")
print(f"REP3: μ={mu_rep3:.4f}, σ={sigma_rep3:.4f}")
print("\nResultados:")
print(df[["ID_deNovo", "Z_Score_Combined", "R_Z_Score"]].head())
```

```python
df_S = DB_DOCK.merge(df, on='ID_deNovo', how='left')
df_S
```

```python
df_S.sort_values(by='R_Z_Score', ascending=True, inplace=True)
df_S
```

##Zscore_LE

```python
df_Rep1_LE = DB_DOCK[['ID_deNovo', 'LE_Rep1']].copy()
df_Rep2_LE = DB_DOCK[['ID_deNovo', 'LE_Rep2']].copy()
df_Rep3_LE = DB_DOCK[['ID_deNovo','LE_Rep3']].copy()
df_Rep1_LE
```

```python
# We extract the relevant columns into an array
Rep1_LE = df_Rep1_LE['LE_Rep1'].values
Rep2_LE = df_Rep2_LE['LE_Rep2'].values
Rep3_LE = df_Rep3_LE['LE_Rep3'].values
```

```python
# Mean and standard deviation for Rep1
mu_rep1_LE = np.mean(Rep1_LE)
sigma_rep1_LE = np.std(Rep1_LE)

# Mean and standard deviation for Rep2
mu_rep2_LE = np.mean(Rep2_LE)
sigma_rep2_LE = np.std(Rep2_LE)

# Mean and standard deviation for Rep3
mu_rep3_LE = np.mean(Rep3_LE)
sigma_rep3_LE = np.std(Rep3_LE)
```

```python
#Z-score_LE

df_LE = DB_DOCK[['ID_deNovo']].copy()

# Z-scores for each method
df_LE['Z_Rep1_LE'] = (df_Rep1_LE['LE_Rep1'] - mu_rep1_LE) / sigma_rep1_LE
df_LE['Z_Rep2_LE'] = (df_Rep2_LE['LE_Rep2'] - mu_rep2_LE) / sigma_rep2_LE
df_LE['Z_Rep3_LE'] = (df_Rep3_LE['LE_Rep3'] - mu_rep3_LE) / sigma_rep3_LE

# Combined Z-score (average of the Z-scores)
df_LE['Z_Score_Combined_LE'] = (df_LE['Z_Rep1_LE'] + df_LE['Z_Rep2_LE'] + df_LE['Z_Rep3_LE']) / 3


# Rank the combined Z-score
df_LE["R_Z_Score_LE"] = df_LE["Z_Score_Combined_LE"].rank(ascending=True, method="dense")

print("Estadísticas calculadas:")
print(f"LE_Rep1: μ={mu_rep1_LE:.4f}, σ={sigma_rep1_LE:.4f}")
print(f"LE_Rep1: μ={mu_rep2_LE:.4f}, σ={sigma_rep2_LE:.4f}")
print(f"LE_Rep1: μ={mu_rep3_LE:.4f}, σ={sigma_rep3_LE:.4f}")
print("\nResultados:")
print(df_LE[["ID_deNovo", "Z_Score_Combined_LE", "R_Z_Score_LE"]].head())
```

```python
columnas_deseadas = ['ID_deNovo', 'SMILES_x', 'SMILES_curated_x', 'Database_x', 'Formula', 'pKd',
       'Scores_Rep1_x', 'Scores_Rep2_x', 'Scores_Rep3_x', 'LE_Rep1_x', 'LE_Rep2_x', 'LE_Rep3_x',
        'Z_Rep1', 'Z_Rep2', 'Z_Rep3',
       'Z_Score_Combined', 'R_Z_Score', 'Z_Rep1_LE', 'Z_Rep2_LE', 'Z_Rep3_LE',
       'Z_Score_Combined_LE', 'R_Z_Score_LE'
      ]

DB_filtrado = DB_denovo_final[columnas_deseadas]
DB_filtrado = DB_denovo_final[columnas_deseadas].rename(columns=lambda x: x.replace('_x',''))
DB_filtrado
```

```python
DB_filtrado.columns
```

```python
# Normalize to a value between 0 and 1
DB_filtrado['Z_S_norm'] = (DB_filtrado['Z_Score_Combined'] - DB_filtrado['Z_Score_Combined'].min()) / (DB_filtrado['Z_Score_Combined'].max() - DB_filtrado['Z_Score_Combined'].min())
DB_filtrado['Z_LE_norm'] = (DB_filtrado['Z_Score_Combined_LE'] - DB_filtrado['Z_Score_Combined_LE'].min()) / (DB_filtrado['Z_Score_Combined_LE'].max() - DB_filtrado['Z_Score_Combined_LE'].min())
DB_filtrado['pKd_norm'] = (DB_filtrado['pKd'] - DB_filtrado['pKd'].min()) / (DB_filtrado['pKd'].max() - DB_filtrado['pKd'].min())

# Average of standardized scores
DB_filtrado['Z_combined_mean'] = (
    DB_filtrado['Z_S_norm'] +
    DB_filtrado['Z_LE_norm'] +
    (1 - DB_filtrado['pKd_norm'])  # We invest to ensure that smaller remains better
) / 3

# Top 5 by combined metric
top_5_combined_describe = DB_filtrado.nsmallest(5, 'Z_combined_mean')
top_5_Z_combined = top_5_combined_describe[['ID_deNovo', 'SMILES_curated', 'Database', 'pKd', 'Z_Score_Combined', 'Z_Score_Combined_LE','Z_S_norm', 'Z_LE_norm', 'pKd_norm', 'Z_combined_mean']]
```
